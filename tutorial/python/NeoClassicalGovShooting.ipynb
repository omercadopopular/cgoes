# A general neoclassical growth model with government

Time is discrete. Households maximize the present discounted value of lifetime utility:

\begin{eqnarray}
    &\max_{\{c_t, k_{t+1} \}_{t\ge0}}& \sum_{t=0}^{\infty} \beta^{t} \mathbb{E}_0 \Bigg[ \frac{c_t^{1-\sigma}}{1-\sigma} + \Gamma(c_t^g) \Bigg] \\
    &s.t.& c_t + k_{t+1} = w_t n_t + (1-\tau_t^k) r_t k_{t} + \tau_t^k \delta k_{t} +  (1-\delta)k_{t} + \varphi_t + \Pi_t, \qquad k_{0} \text{ given}, (n_t = 1 \forall t) \nonumber
\end{eqnarray}

We can write the Lagrangian for this problem:

\begin{equation*}
    \mathcal{L} = \mathbb{E}_0 \sum_{t=0}^{\infty} \beta^{t} \Bigg[   \frac{c_t^{1-\sigma}}{1-\sigma} + \Gamma(g_t^c) + \lambda_t \Big( w_t + (1-\tau_t^k) r_t k_{t} + \tau_t^k \delta k_{t} + \varphi_t + \Pi_t - c_t - k_{t+1} + (1-\delta)k_{t} \Big) \Bigg]
\end{equation*}

 whose first order conditions satisfy, for every $t \ge 0$:

\begin{eqnarray*}
    c_t &:& c_t^{-\sigma} = \lambda_t \\
    k_{t+1} &:& \lambda_t = \beta \lambda_{t+1} \Big[ r_{t+1} -  \tau_{t+1}^k ( r_{t+1} - \delta) + (1-\delta) \Big]
\end{eqnarray*}

From the first order conditions, we can derive Euler Equation:

\begin{eqnarray}
    \Big( \frac{c_{t+1}}{c_{t}} \Big)^{\sigma} = \beta \Big[ r_{t+1} -  \tau_{t+1}^k ( r_{t+1} - \delta) + (1-\delta) \Big]
\end{eqnarray}

 which shows that capital taxes distort savings (i.e., future consumption) downwards, under the restriction that $r_{t+1} > \delta$ (otherwise net capital income is negative and the tax becomes a subsidy). There is a representative firm endowed with a constant-returns to scale technology that maximizes the following objective function:

\begin{eqnarray}
    \pi_t(z_t) = \max_{\{\tilde{k}_{t}, \tilde{n}_t\}_{t\ge0}}   y_t - r_t \tilde{k}_{t} - w_t \tilde{n}_t, \qquad s.t. \quad y_t(z_t) = z_t \tilde{k}_{t}^{\alpha} \tilde{n}_t^{1-\alpha} 
\end{eqnarray}

where $z_t$ may be stochastic or deterministic. If an equilibrium exists, factor prices will equal their marginal products:

\begin{eqnarray}
\alpha \frac{y_t}{\tilde{k}_{t}} &=& r_t \\
(1-\alpha) \frac{y_t}{\tilde{n}_{t}} &=& w_t 
\end{eqnarray}

Finally, there is a government that collects labor, capital, and lump-sum taxes ($\varphi_t < 0$) and spends them on government consumption, government investment or lump-sum transfers ($\varphi_t > 0$). We assume that the government's budget balances for every period $t\ge0$, i.e.:

\begin{equation}
    c_t^g + \varphi_t = \tau_t^k (r_t - \delta) k_{t}  
\end{equation}

A **competitive equilibrium** with government consumption is characterized by a sequence of household allocations $\{c_t, k_{t+1}, n_t\}_{t\ge0}$, with $k_{0}$ given; a sequence of firms allocations for capital and labor demand $\{\tilde{k}_t, \tilde{n}_t\}_{t\ge0}$; a sequence of government consumption, taxes, and transfers $\{c^g_t, \tau_t^k, \varphi_t \}_{t\ge0}$; and a sequence of factor prices $\{w_t, r_t\}_{t\ge0}$, such that:

- given prices, household allocations solve the household's problem for every $t\ge0$
- given prices, firm allocations solve the firm's problem for every $t\ge0$;
- the government budget balances for every $t\ge0$;
- factor markets clear ($k_{t}= \tilde{k}_{t}$ and $n_{t}= \tilde{n}_{t}$) for every $t\ge0$;
- due to free entry, profits are zero ($y_t = r_t k_{t} + w_t n_t$)  for every $t\ge0$; and
- goods markets clear ($c_t + i_t + c_t^g  = y_t$)  for every $t\ge0$.

# Characterizing the Steady-State

Suppose that steady state that $z_t = z \forall t$ and steady-state government consumption to income ratio $\frac{c^g}{y} = \kappa^c$ is a constant. Let us solve for the steady state of this economy. We drop the time subscripts to indicate steady state values. Using the labor-leisure condition and the firm's labor demand:

\begin{eqnarray*}
     r &=& \frac{1 - \beta[ 1 - (1-\tau^k) \delta]}{  (1-\tau^k) \beta } \qquad (\text{from euler eqn}) \\
     \frac{k}{y} &=& \frac{\alpha}{r} \qquad (\text{from capital demand}) \\
    \frac{i}{y} &=& \delta \frac{k}{y} \qquad (\text{from capital accumulation}) \\
    \frac{c^g}{y} &=& \kappa^c   \qquad (\text{exogenous}) \\
    \frac{\varphi}{y} &=& \tau^k (r - \delta) \frac{k}{y} - \frac{c^g}{y}  \qquad (\text{from govt budget cnst})  \\
    \frac{c}{y} &=& 1 - \frac{c^g}{y} - \frac{i}{y} \qquad (\text{from good markets clearance}) \\
    n &=& 1 \qquad (\text{assumption}) \\
    y &=& \Big[ z (k/y)^{\alpha} \Big]^{\frac{1}{1-\alpha}} \qquad (\text{from production function}) \\
    w &=& (1-\alpha) y \qquad (\text{from labor demand}) \\ 
    k &=& \frac{k}{y} \times y = k/y \left[ z \left( k/y \right)^{\alpha} \right]^{\frac{1}{1-\alpha}} = \left[ z k/y \right]^{\frac{1}{1-\alpha}} = \left[ \frac{  z \alpha (1-\tau^k) \beta }{1 - \beta[ 1 - (1-\tau^k) \delta]} \right]^{\frac{1}{1-\alpha}}
\end{eqnarray*}

# Simplifying things for an initial shooting algorithm

You will realize that this model is pretty general. This is useful but might get in the way of learning these techniques. For now, let us assume that households supply labor inellastically as $n_t = 1 \forall t$. Of course, in that case the labor leisure condition becomes irrelevant, wages are always determined as $w_t = (1-\alpha)y_t$ and the government can always capture total labor income $nw=w=(1-\alpha)y$ (why?). We will get back to the general model later.

Given that $n_t = 1 \forall t$, our goal is to solve for the **competitive equilibrium** with government consumption. Looking back at the definition, this is characterized by a sequence of household allocations $\{c_t, k_{t+1}, n_t=1\}_{t\ge0}$, with $k_{0}$ given; a sequence of firms allocations for capital and labor demand such that the factor markets clear $\{\tilde{k}_t=k_t, \tilde{n}_t=1\}_{t\ge0}$; a sequence of factor prices $\{w_t, r_t\}_{t\ge0}$; and an exogenously given sequence of government consumption, taxes, and transfers $\{c^g_t, \tau_t^n, \tau_t^k, \varphi_t \}_{t\ge0}$ that we will make sure respects he government budget constraint.

We will use some initial guess $c_0^{m}$ and the relevant equilibrium conditions solve for a path of consumption choices. First, combining the euler equation and the equilibrium condition for rents:

\begin{eqnarray*}
    c_{t+1} &=& c_{t} \left( \beta \Big[ (1-\tau_{t+1}^k)  r_{t+1} + (1-\delta) + \delta \tau_{t+1}^k \Big] \right)^{1/\sigma}, \qquad r_{t+1} = \alpha \frac{y_{t+1}}{k_{t+1}} = \alpha k_{t+1}^{\alpha-1} \\
    \implies c_{t+1} &=& c_{t} \left( \beta \Big[ (1-\tau_{t+1}^k)  \alpha k_{t+1}^{\alpha-1} + (1-\delta) + \delta \tau_{t+1}^k \Big] \right)^{1/\sigma}
\end{eqnarray*}

Then, using the law of motion for capital acumulation:

\begin{eqnarray*}
    k_{t+1} &=& w_t + (1-\tau_t^k) r_t k_{t} + \tau_t^k \delta k_{t} +  (1-\delta)k_{t} + \varphi_t - c_t \\
    k_{t+1} &=& (1-\alpha) k_t^{\alpha} + (1-\tau_t^k) \alpha k_t^{\alpha} + \tau_t^k \delta k_{t} +  (1-\delta)k_{t} + \varphi_t - c_t \\
    k_{t+1} &=& \left( (1-\alpha) + (1-\tau_t^k) \alpha \right) k_t^{\alpha} + \tau_t^k \delta k_{t} +  (1-\delta)k_{t} + \varphi_t - c_t
\end{eqnarray*}


Assuming that when $t \rightarrow \infty$, the fiscal policy becomes constant, we can check if the economy is reaching the terminal steady state.  The shooting algorithm works as follows:


1. Choose some max number of iterations $M$.
2. Choose a small tolerance factor $\varepsilon$.
3. Solve the model for the terminal steady state $k_{ss}$.
4. Select a large time index $T$, such that the economy reaches a steady state after the full transition.
5. Guess an initial jump variable $c_0^m$ for iteration $m=0$
6. For iteration $m$:
    - If $m > M$, stop. Your algorithm failed to converge.
    - Find $k_1^m$ by using $c_0^m$ and $k_0^m$ in the resource constraint.
    - Find $c_1^m$ by using $k_1^m$ and $c_0^m$ in the Euler equation. 
    - Proceed by using  $c^m_t$ and $k^m_{t+1}$ to find $c^m_{t+1}$ and $k^m_{t+1}$ up until $c^m_{T}$ and $k^m_{T}$.
    - If $\hat{k}_{T} \approx k_{ss}$, i.e. $|\hat{k}^m_{T} - k_{ss}| \le \varepsilon$, stop. Your algorithm converged.
    - If $k^m_{T} - k_{ss} > \varepsilon$ , set $m = m + 1$, raise the initial guess $c_{0}^m$ and go back to step 6.
    - If $k^m_{T} - k_{ss} < -\varepsilon$, set $m = m + 1$, lower the initial guess $c_0^m$ and go back to step 6.

**Analogy**: If the class is the blueprint of a car, an instance is an actual car that has been built based on that blueprint. You can have multiple cars (instances) that are based on the same blueprint (class), but each car can have different details (e.g., different makes and models).

# Back to economics

In our case, we will create a `class`: `gmGov` (for neoclassical growth model with a government). This `class` is a blueprint, so it can describe a model with any possible values of the starting value of $k_0$, parameters ($\alpha, \beta, \sigma, \delta$), streams of government variables ($\{ \tau_t^k, c^g_t \}_{t\ge 0}$, with $\varphi_t = \tau_t^k (r_t - \delta) k_{t} - c_t^g$ omitted since it can be changed to ensure the budget balances), technological shocks ($\{ z_t \}_{t\ge 0}$), as well as some specific inputs for the numerical simulation.

```python
import numpy as np
import matplotlib.pyplot as plt

class gmGov:

    def __init__(self, initial, params, gov, tech, T=1000, M = 10000, ε = 10**(-3)):
        self.initial = initial
        self.params = params
        self.gov = gov
        self.tech = tech

        self.k0 = initial
        self.α, self.β, self.σ, self.δ  =  params
        self.τk, self.cg = gov
        self.z = tech

        self.T, self.M, self.ε = T, M, ε
        
        self.ss()
    
    def ss(self):
        self.τkss = self.τk[-1]
        self.cgss = self.cg[-1]
        self.zss = self.z[-1]
        self.nss = 1
        rss = (1 - self.β * (1-(1-self.τkss)* self.δ) ) / ( (1-self.τkss) * self.β )
        kyss = self.α/rss
        self.yss = ( self.zss * kyss**(self.α) * self.nss**(1-self.α) )**(1/(1-self.α))
        self.kss = kyss * self.yss
        self.css = self.yss - self.δ * self.kss - self.cgss

    def shootGuess(self,c0,k0):
        
        k = np.zeros(self.T+1)
        c = np.zeros(self.T+1)
        y = np.zeros(self.T+1)       
        φ = np.zeros(self.T+1)

        c[0], k[0] = c0, k0
       
        α, β, σ, δ = self.params
        τk, cg = self.gov
        z = self.tech
            
        for t in range(T):

            # assure that the government budget balances
            φ[t] = τk[t] * ( α * k[t]**α - δ * k[t] ) - cg[t]  

            # production function
            y[t] = z[t] * k[t]**α # production function

            # update path of consumption and savings
            k[t+1] =  y[t] * ( (1-α) + (1-τk[t])*α ) + τk[t] * δ * k[t] + (1-δ)*k[t] + φ[t] - c[t]
            if k[t+1]<0:
                k[t+1] = 0.0001 # non-negative constraint
            c[t+1] = c[t] * ( β * ( (1-τk[t+1])*α*z[t+1]*k[t+1]**(α-1) + (1-δ) + δ * τk[t+1] )  )**(1/σ)
    
        return k, c
    
    def boundsInitialize(self):
        c_lower = self.css/30
        
        k, c = self.shootGuess(c_lower,self.k0)
        if (k[-1] - self.kss)/self.kss < 0:
            print("lower bound too high")
        self.c_lower = c_lower
        
        c_upper = self.css*10
        k, c = self.shootGuess(c_upper,self.k0)
        if (k[-1] - self.kss)/self.kss > 0:
            print("upper bound too low")
        self.c_upper = c_upper
    
        self.c0 = (c_lower+c_upper)/2 # initial guess
    
    def shootConvergence(self):
        self.boundsInitialize()
        
        m = 0
        while m <= self.M:        
            k, c = self.shootGuess(self.c0,self.k0)
    
            dist = (k[-1] - self.kss)/self.kss # distance in % from the SS
            if abs(dist) < self.ε: # if dist smaller than tolerance we are done
                print("Iteration Finished: {}, Distance: {}".format(m,dist))
                break
    
    		# update guess
            if dist > 0:
                self.c_lower = self.c0
                self.c0 = (self.c0 + self.c_upper)/2
            elif dist < 0:
                self.c_upper = self.c0
                self.c0 = (self.c0 + self.c_lower)/2
    
            m += 1
        
        y = self.z * k**self.α
        i = (y - c)
        s = i/y
                
        return k, c, y, i, s
    
    def plotPath(self):
        k, c, y, i, s = self.shootConvergence()
        
        fig, (ax1) = plt.subplots(1, 1, figsize=(14, 14))

        ax1.plot(np.linspace(1,T+1,T+1), k, label='$K_t$', color='blue')
        ax1.plot(np.linspace(1,T+1,T+1), c, label='$C_t$', color='red')
        ax1.plot(np.linspace(1,T+1,T+1), y, label='$Y_t$', color='purple')
        ax1.plot(np.linspace(1,T+1,T+1), i, label='$I_t$', color='black')
        ax1.set_title('Convergence')
        ax1.set_xlabel('$t$')
        ax1.legend()
        
        # Adjust layout
        plt.tight_layout()
        
        # Show the plot
        plt.show()
```

If we create a specific model solution or simulation (an `instance`), we need to specify those inputs. That `instance` will have methods. For instance, the `method` `plotPath()` allows us to plot the convergence towards the steady state.

```python
# Set the parameters values
α = 0.33
β = 0.95
σ = 2.0
δ = 0.2
T = 100
τk = np.zeros(T+1)
z = np.ones(T+1)
cg = np.zeros(T+1)

initial = (10)
params = (α,β,σ,δ)
gov  = (τk,cg)
tech = (z)

sim1 = gmGov(initial,params,gov,tech,T)

sim1.plotPath()
```

# Different Kinds of Government Financing

## What happens if government expenditure increases?

Note that the model above nests the simple neoclassical growth model we have seen if we set $\tau^k=\eta_t=g_t^c=0 (\forall t)$. Suppose that government consumption increases permanently to some positive number. The government has three instruments to finance the new expenditures: taxes on capital; and lump-sum taxes.

Now let us do some comparative statics. First, it is easy to see that $r$ is increasing in $\tau^k$:

\begin{equation*}
    \frac{\partial r}{\partial \tau^k} = \frac{\partial}{\partial \tau^k}  \Big[ \frac{1-\beta}{\beta(1-\tau^k)}  - \delta \Big] = \frac{1-\beta}{\beta(1-\tau^k)^2} > 0
\end{equation*}

From this, we can see that steady-state capital stock is decreasing in $\tau^k$:

\begin{equation*} 
    \frac{\partial}{\partial \tau^k} \frac{k}{y} = \frac{\partial}{\partial r} \frac{\alpha}{r}  \frac{\partial r}{\partial \tau^k} = -\frac{\alpha}{r^2} \frac{\partial r}{\partial \tau^k} < 0
\end{equation*}

Therefore, total wealth as a share of income in this economy ($k/y$) decreases if capital taxes increase. Realizing that steady state output $y$ is increasing in $k/y$ leads to the conclusion that $y$ is decreasing in $\tau^k$.

Unlike taxes on capital, lump-sum taxes do not create wedges on the optimality conditions. This does not mean, however, that nothing changes. Note that, at the steady-state:

\begin{equation*}
    \frac{c}{y} = 1 - \frac{c^g}{y} - \frac{i}{y}
\end{equation*}

which implies that $c/y$ decreases with a positive $c^g$.

Now let us take that to our numerical model. Suppose that we start at the steady state of the model, but government expenditures increase. What will happen next will depend on how the government is financing itself.

```python
# Set the parameters values
α = 0.33
β = 0.95
σ = 2.0
δ = 0.2
T = 100
τk = np.zeros(T+1)
z = np.ones(T+1)
cg = np.zeros(T+1)
cg[10:] = 0.1

initial = (sim1.kss)
params = (α,β,σ,δ)
gov = (τk,cg)
tech = (z)

sim2 = gmGov(initial,params,gov,tech,T)

sim2.plotPath()
```

# Adding Labor
We will now add an endogenous labor supply to our model. It may seem repetitive, but we will describe the whole model again, including optimality conditions and steady state solutions.

Time is discrete. Households maximize the present discounted value of lifetime utility:

\begin{eqnarray}
    &\max_{\{c_t, k_{t+1}, n_t \}_{t\ge0}}& \sum_{t=0}^{\infty} \beta^{t} \mathbb{E}_0 \Bigg[ \frac{c_t^{1-\sigma}}{1-\sigma} - \psi \frac{n_t^{1+\eta}}{1+\eta}  + \Gamma(c_t^g) \Bigg] \\
    &s.t.& c_t + k_{t+1} = (1-\tau_t^n)w_t n_t + (1-\tau_t^k) r_t k_{t} + \tau_t^k \delta k_{t} +  (1-\delta)k_{t} + \varphi_t + \Pi_t, \qquad k_{0} \text{ given} \nonumber
\end{eqnarray}

We can write the Lagrangian for this problem:

\begin{equation*}
    \mathcal{L} = \mathbb{E}_0 \sum_{t=0}^{\infty} \beta^{t} \Bigg[   \frac{c_t^{1-\sigma}}{1-\sigma} - \psi \frac{n_t^{1+\eta}}{1+\eta}  + \Gamma(g_t^c) + \lambda_t \Big( (1-\tau_t^n)w_t n_t + (1-\tau_t^k) r_t k_{t} + \tau_t^k \delta k_{t} + \varphi_t + \Pi_t - c_t - k_{t+1} + (1-\delta)k_{t} \Big) \Bigg]
\end{equation*}

 whose first order conditions satisfy, for every $t \ge 0$:

\begin{eqnarray*}
    c_t &:& c_t^{-\sigma} = \lambda_t \\
    n_t &:& \psi n_t^{\eta} = \lambda_t (1-\tau_t^n)w_t \\
    k_{t+1} &:& \lambda_t = \beta \lambda_{t+1} \Big[ r_{t+1} -  \tau_{t+1}^k ( r_{t+1} - \delta) + (1-\delta) \Big]
\end{eqnarray*}

From the first order conditions, we can derive the labor-leisure condition:

\begin{equation}
    \psi n_t^{\eta} = c_t^{-\sigma} (1-\tau_t^n)w_t
\end{equation}

 which shows that labor taxes distort labor supply downwards. We can also derive Euler Equation:

\begin{eqnarray}
    \Big( \frac{c_{t+1}}{c_{t}} \Big)^{\sigma} = \beta \Big[ r_{t+1} -  \tau_{t+1}^k ( r_{t+1} - \delta) + (1-\delta) \Big]
\end{eqnarray}

 which shows that capital taxes distort savings (i.e., future consumption) downwards, under the restriction that $r_{t+1} > \delta$.

There is a representative firm endowed with a constant-returns to scale technology that maximizes the following objective function:

\begin{eqnarray}
    \pi_t(z_t) = \max_{\{\tilde{k}_{t}, \tilde{n}_t\}_{t\ge0}}   y_t - r_t \tilde{k}_{t} - w_t n_t, \qquad s.t. \quad y_t(z_t) = z_t \tilde{k}_{t}^{\alpha} \tilde{n}_t^{1-\alpha} 
\end{eqnarray}

where $z_t$ may be stochastic or deterministic. If an equilibrium exists, factor prices will equal their marginal products:

\begin{eqnarray}
\alpha \frac{y_t}{\tilde{k}_{t}} &=& r_t \\
(1-\alpha) \frac{y_t}{\tilde{n}_{t}} &=& w_t 
\end{eqnarray}

Finally, there is a government that collects labor, capital, and lump-sum taxes ($\varphi_t < 0$) and spends them on government consumption, government investment or lump-sum transfers ($\varphi_t > 0$). We assume that the government's budget balances for every period $t\ge0$, i.e.:

\begin{equation}
    c_t^g + \varphi_t = \tau_t^n w_t n_t + \tau_t^k (r_t - \delta) k_{t}  
\end{equation}

Suppose that steady state that $z_t = z \forall t$ and steady-state government consumption to income ratio $\frac{c^g}{y} = \kappa^c$ is a constant. Let us solve for the steady state of this economy. We drop the time subscripts to indicate steady state values. Using the labor-leisure condition and the firm's labor demand:

\begin{eqnarray*}
     r &=& \frac{1 - \beta[ 1 - (1-\tau^k) \delta]}{  (1-\tau^k) \beta } \qquad (\text{from euler eqn}) \\
     \frac{k}{y} &=& \frac{\alpha}{r} \qquad (\text{from capital demand}) \\

    \frac{c^g}{y} &=& \kappa^c   \qquad (\text{exogenous}) \\
    \frac{\varphi}{y} &=& (1-\alpha) \tau^n + \tau^k (r - \delta) \frac{k}{y} - \frac{c^g}{y}  \qquad (\text{from labor demand + govt budget cnst})  \\
    \frac{c}{y} &=& 1 - \frac{c^g}{y} - \frac{i}{y} \qquad (\text{from good markets clearance}) \\
    n &=& \Big[ \frac{1}{\psi} (c/y)^{-\sigma} (1-\tau^n) (1-\alpha) \Big]^{\frac{1}{ \eta + \sigma}} \Big[ z (k/y)^{\alpha} \Big]^{\frac{1-\sigma}{(1-\alpha)(\eta+\sigma)}} \qquad (\text{see Appendix}) \\
    y &=& \Big[ z (k/y)^{\alpha} n^{1-\alpha} \Big]^{\frac{1}{1-\alpha}} \qquad (\text{from production function}) \\
    w &=& (1-\alpha) \frac{y}{n} \qquad (\text{from labor demand})
\end{eqnarray*}

# Changes needed in the code

There are some small changes needed in the code and one quite important one. Small ones are that labor income shows up in the household and government budget constraints. But this is not a huge problem, since we can use the labor leisure condition to express $n_{t}$ as a function of $c_{t}$ choices, and we have already designed an algorithm to solve for the optimal $c_{t}$:

\begin{equation*}
    n_t = \left( \frac{1}{\psi} \times (1-\alpha) z_t k_t^{\alpha} \times c_{t}^{-\sigma} \right)^{1/(\eta+\alpha)}
\end{equation*}

A more difficult problem is solving for future consumption $c_{t+1}$ using the euler equation. The problem is that $c_{t+1}$ depends on future interst rates $r_{t+1}$, which in turn depends on optimal choices of labor $n_{t+1}$. In short, $c_{t+1}$ depends on $n_{t+1}$ but $n_{t+1}$ also depends on $c_{t+1}$.

So we will proceed in the following way. We will include a new `method` in our `class` that, given some $k_{t+1}$ and all exogenous paths and parameters,  will:

- Guess some value for $n_{t+1}$ we will use the initial guess $n^j_{t+1}=n_{t+1}$
- Calculates $c^j_{t+1} = c_t \left( \beta \times \left[ (1-\tau^k_t) \alpha z_{t+1} \left( \frac{n^j_{t+1}}{k_{t+1}} \right)^{1-\alpha} + (1-\delta) + \delta \tau^k_{t+1} \right] \right)$
- Update your guess using: $n^{j+1}_{t+1} = \left( \frac{1}{\psi} \times (1-\alpha) z_{t+1} k_{t+1}^{\alpha} \times \left( c^{j}_{t+1}\right)^{-\sigma} \right)^{1/(\eta+\alpha)}$
- If $|n^{j+1}_{t+1}-n^j_{t+1}|/n^j_{t+1} < \varepsilon$, stop.


```python
import numpy as np
import matplotlib.pyplot as plt

class gmGovL:

    def __init__(self, initial, params, gov, tech, T=1000, M = 10000, ε = 10**(-3)):
        self.initial = initial
        self.params = params
        self.gov = gov
        self.tech = tech

        self.k0 = initial
        self.α, self.β, self.σ, self.δ, self.ψ, self.η  =  params
        self.τk, self.τn, self.cg, self.cgyss = gov
        self.z = tech

        self.T, self.M, self.ε = T, M, ε
        
        self.ss()
    
    def ss(self):
        self.τkss, self.τnss, self.zss = self.τk[-1], self.τn[-1], self.z[-1]
        self.rss = (1 - self.β * (1-(1-self.τkss)* self.δ) ) / ( (1-self.τkss) * self.β )
        self.kyss = self.α / self.rss
        self.iyss = self.δ * self.kyss
        self.cyss = 1 - self.iyss - self.cgyss
        self.φyss = (1-self.α) * self.τnss + self.τkss * (self.rss - self.δ) * self.kyss - self.cgyss
        self.nss = (1/self.ψ * self.cyss**(-self.σ) * (1-self.τnss) * (1-self.α) )**(1/(self.η + self.α)) * (self.zss * self.kyss**self.α )**((1-self.σ)/((1-self.α)*(self.η+self.α)))
        self.yss = ( self.zss * self.kyss**(self.α) * self.nss**(1-self.α) )**(1/(1-self.α))
        self.kss = self.kyss * self.yss
        self.wss = (1-self.α) * self.yss / self.nss
        self.css = self.cyss * self.yss

    def laborLeisureGuess(self,c0,n0,k1,t,M=1000, ε=10**(-3)):
        α, β, σ, δ, ψ, η  = self.params
        z = self.z
        τk, τn = self.τk, self.τn

        cPath = np.zeros(M+1)
        nPath = np.zeros(M+1)

        # first guesses
        nPath[0] = n0
        cPath[0] = c0 * ( β * ( (1-τk[t])*α*z[t]*k1**(α-1)*nPath[0]**(1-α) + (1-δ) + δ * τk[t] )  )**(1/σ)

        for j in range(M):
            cPath[j+1] = c0 * ( β * ( (1-τk[t+1])*α*z[t+1]*k1**(α-1)*nPath[j]**(1-α) + (1-δ) + δ * τk[t+1] )  )**(1/σ)
            nPath[j+1] = (cPath[j+1]**(-σ) * (1-τn[t+1]) * z[t+1] * k1**α * (1-α) / ψ)**(1/(η+α))
        
            dist = abs(nPath[j+1] - nPath[j])/nPath[j]
            if dist < 0.001:
                break

        return cPath[j+1], nPath[j+1]
        
                
    def shootGuess(self,c0,k0):
        
        k = np.zeros(self.T+1)
        c = np.zeros(self.T+1)
        n = np.zeros(self.T+1)
        y = np.zeros(self.T+1)       
        φ = np.zeros(self.T+1)
        
        α, β, σ, δ, ψ, η  = self.params
        τk, τn, cg, cgyss = self.gov
        z = self.z
        
        c[0], k[0] = c0, k0
        n[0] = (c[0]**(-σ) * (1-τn[0]) * z[0] * k[0]**α * (1-α) / self.ψ)**(1/(η+α)) # labor-leisure condition
           
        for t in range(T):
            # assure that the government budget balances
            φ[t] = τk[t] * ( α * k[t]**α - δ * k[t] ) - cg[t]  

            # production function
            y[t] = z[t] * k[t]**α * n[t]**(1-α) # production function

            # update path of savings
            k[t+1] =  y[t] * ( (1-τn[t])*(1-α) + (1-τk[t])*α ) + τk[t] * δ * k[t] + (1-δ)*k[t] + φ[t] - c[t]
            if k[t+1]<0:
                k[t+1] = 0.0001 # non-negative constraint
                
            # update path of consumption and labor
            c[t+1], n[t+1] = self.laborLeisureGuess(c[t],n[t],k[t+1],t)
    
        return k, c, n

    def boundsInitialize(self):
        c_lower = self.css/30
        
        k, c, n = self.shootGuess(c_lower,self.k0)
        if (k[-1] - self.kss)/self.kss < 0:
            print("lower bound too high")
        self.c_lower = c_lower
        
        c_upper = self.css*10
        k, c, n = self.shootGuess(c_upper,self.k0)
        if (k[-1] - self.kss)/self.kss > 0:
            print("upper bound too low")
        self.c_upper = c_upper
    
        self.c0 = (c_lower+c_upper)/2 # initial guess
    
    def shootConvergence(self):
        self.boundsInitialize()
        
        m = 0
        while m <= self.M:        
            k, c, n = self.shootGuess(self.c0,self.k0)
    
            dist = (k[-1] - self.kss)/self.kss # distance in % from the SS
            if abs(dist) < self.ε: # if dist smaller than tolerance we are done
                print("Iteration Finished: {}, Distance: {}".format(m,dist))
                break
    
    		# update guess
            if dist > 0:
                self.c_lower = self.c0
                self.c0 = (self.c0 + self.c_upper)/2
            elif dist < 0:
                self.c_upper = self.c0
                self.c0 = (self.c0 + self.c_lower)/2
    
            m += 1
        
        y = self.z * k**self.α
        i = (y - c)
        s = i/y
                
        return k, c, y, i, s, n
    
    def plotPath(self):
        k, c, y, i, s, n = self.shootConvergence()
        
        fig, (ax1) = plt.subplots(1, 1, figsize=(14, 14))

        ax1.plot(np.linspace(1,T+1,T+1), k, label='$K_t$', color='blue')
        ax1.plot(np.linspace(1,T+1,T+1), c, label='$C_t$', color='red')
        ax1.plot(np.linspace(1,T+1,T+1), y, label='$Y_t$', color='purple')
        ax1.plot(np.linspace(1,T+1,T+1), i, label='$I_t$', color='black')
        ax1.plot(np.linspace(1,T+1,T+1), n, label='$N_t$', color='yellow')
        ax1.set_title('Convergence')
        ax1.set_xlabel('$t$')
        ax1.legend()
        
        # Adjust layout
        plt.tight_layout()
        
        # Show the plot
        plt.show()
```

```python
# Set the parameters values
α = 0.33
β = 0.95
σ = 2.0
δ = 0.2
ψ = 1
η = 3
T = 50
τk = np.zeros(T+1)
τn = np.zeros(T+1)
z = np.ones(T+1)
cg = np.zeros(T+1)
cgyss = 0

initial = (10)
params = (α,β,σ,δ,ψ,η)
gov  = (τk,τn,cg,cgyss)
tech = (z)

sim4 = gmGovL(initial,params,gov,tech,T)
```
